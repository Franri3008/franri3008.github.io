```python
import pandas as pd
df = pd.read_csv("x_y_z.csv");
colors = {"A": "#448AFF", "B": "#1565C0", "C": "#009688", "D": "#8BC34A", "E": "#FFC107", "F": "#FF9800", "G": "#F44336", "H": "#AD1457", "Y": "#565656"};
df["letter"] = df["node"].str[0];
df["color"] = df["letter"].apply(lambda x: colors.get(x, "#FFFFFF"));
df.drop(columns=["letter"], inplace=True);
df.to_csv("x_y_z.csv", index=False);
df.head()
```
