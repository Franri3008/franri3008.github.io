```python
import pandas as pd
```

```python
df = pd.read_csv("YoungClimateLeaders.csv", on_bad_lines="skip");
df.head()
```

```python
from CustomTools.llm import embedding

to_emb = df[["Id", "What do you think is the biggest challenge for climate action today? (Up to 250 words)"]];
to_emb.rename(columns={"Id": "id", "What do you think is the biggest challenge for climate action today? (Up to 250 words)": "text"}, inplace=True);

emb = embedding.embed(to_emb);
emb.head()
```

```python
ai = pd.DataFrame([["ai", "The biggest challenge for climate action today is the gap between commitments and implementation. Most governments and companies have announced ambitious targets for 2030 and 2050, but the policies, investments, and enforcement mechanisms needed to achieve them are not being deployed at the required scale or speed. This gap stems from three intertwined issues. First, political and economic inertia: short-term costs of decarbonization often deter leaders focused on election cycles or quarterly results. Second, insufficient infrastructure and financing in key sectors such as energy, transport, and industrial production, particularly in developing economies that face higher capital costs. Third, fragmented global coordination: major emitters pursue divergent strategies, and geopolitical tensions slow cooperation on technology, standards, and climate finance. Scientific understanding, public awareness, and available clean technologies are no longer the limiting factors. The core obstacle is turning existing knowledge and pledges into sustained, large-scale action."]], columns=["id", "text"]);
ai_emb = embedding.embed(ai);
```

```python
ai_emb.head()
```

```python
is_ai = embedding.cosine_similarity(emb, ai_emb);
is_ai.head()
```

## Dimensionality reduction

```python
import umap
import numpy as np

features = [c for c in emb.columns if c.startswith("d")];
X = emb[features].to_numpy();
reducer = umap.UMAP(n_components=3, random_state=258);

X_reduced = reducer.fit_transform(X);

emb_umap = emb[["id", "text"]];
emb_umap[["u0", "u1", "u2"]] = X_reduced;

emb_umap.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler();
emb_umap[["u0", "u1", "u2"]] = scaler.fit_transform(emb_umap[["u0", "u1", "u2"]]);

emb_umap.head()
```

```python
emb_umap["ai_alert"] = emb_umap["id"].map(is_ai.set_index("from")["similarity"]);
emb_umap.head()
```

```python
def assign_color(row):
    alert = row["ai_alert"];
    if alert <= 0.40:
        return "#45ce1c"
    elif alert <= 0.55:
        return "#bbdb43"
    elif alert <= 0.70:
        return "#f7e379"
    elif alert <= 0.85:
        return "#f3a134"
    return "#e51f1f"

emb_umap["color"] = emb_umap.apply(assign_color, axis=1);
emb_umap.head()
```

```python
emb_umap.to_csv("data.csv", index=False);
```
